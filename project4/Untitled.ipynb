```python
import scipy.io as sio
import numpy as np
bostonData = sio.loadmat('./boston.mat')
```

```python
xTr = bostonData['xTr']
xTe = bostonData['xTe']
```

```python
d, n = np.shape(xTr)
D, N = np.shape(xTe)
## << Remove 2 lines above and insert your solution here

m = np.mean(xTr,axis=1)
M = np.mat(m)
u = np.diag(1/np.std(xTr.T,axis=0))
xTr = u.dot((xTr-np.tile(M.T,n)))
xTe = u.dot((xTe-np.tile(M.T,N)))
xTr = np.array(xTr)
xTe = np.array(xTe)
```

```python
print(xTr)
print(xTe)
print(u)
print(m)
```
